## Loop through a dataset of RGB images, convert to grayscale and save

```python
# path_source = "/scratch/data/images" # folder to READ from
path_source = "/scratch/data/images"
path_dest = "/shared/davidsalvador/data/images_gray" # folder to WRITE to
```

```python
if path_source == path_dest:
    raise Exception("Source Path and Destination Path cannot be the same")
```

```python
my_files = os.listdir(path_source) # list of ALL files in path_source
tif_files = [x for x in my_files if '.tif' in x] # grab only .tif files
```

```python
for file in tif_files:
    image = imread( os.path.join(path_source, file) ) # read rgb image
    image_gray = color.rgb2gray(image) # convert to grayscale
    imsave( os.path.join(path_dest, file), image_gray ) # save grayscale image
```

## Check for successful image conversion

```python
ind = 20000
image = imread( os.path.join(path_source, tif_files[ind]) )
image_gray = imread( os.path.join(path_dest, tif_files[ind]) )
```
